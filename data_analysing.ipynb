```python
ev_county_distribution=df['County'].value_counts()
ev_county_distribution
```

```python
top_county=ev_county_distribution.head(3).index
top_county
```

```python
df['County'].isin(top_county)
```

```python
top_county_data=df[df['County'].isin(top_county)]
```

```python
ev_city_distribution_top_county= top_county_data.groupby(['County','City']).size().sort_values(ascending=False).reset_index(name='No.of vehicles')
```

```python
ev_city_distribution_top_county
```

```python
top_cities=ev_city_distribution_top_county.head(20)
top_cities
```

```python
plt.figure(figsize=(9,6))
sns.barplot(data=top_cities,x='City',y='No.of vehicles',hue='County',palette='magma')
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Electric Range'],bins=30,kde=True)
plt.title('Distribution of electric range')
plt.ylabel('No. of vehicles')
plt.axvline(df['Electric Range'].mean(),color='red',linestyle='--');
```

```python
avg_range_by_year=df.groupby('Model Year')['Electric Range'].mean().reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Model Year',y='Electric Range',data=avg_range_by_year,marker='o')
plt.title('Average electric range by year')
plt.grid(True);
```
